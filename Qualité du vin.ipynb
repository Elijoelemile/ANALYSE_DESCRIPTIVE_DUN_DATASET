# Analyse descriptive de la base de donnée
     
#### Dans cette analyse on va se concentrer sur le résumé descriptif des variables de notre Dataset.

Pourrais-je faire de l'analyse en ayant des données Manquantes, Duppliquées et Aberrantes? Bien sûr et c'est ce que nous allons faire.

Nous allons utiliser:

###### 1.Les Mesures de tendance centrale telles que: <font color="red">la moyenne, la médiane et le mode.</font>

###### 2.Les Mesures de dispersion telles que: <font color="red">l'étendue, variance, écart-type, intervalle interquartile (IQR) .</font>

###### 3.Les Mesures de position telles que: <font color="red">les quartiles et les percentiles.</font>

###### 4.Les Mesures de forme telles que: <font color="red">l'assymétrie(Skewness) et l'aplatissement(Kurtosis).</font>

##### Sujet: On parle du <font color="red">vin.</font> Toutes les analyses faîtes seront en fonction du <font color="red">vin.</font> 

### <font color="green">Résumé descriptif de notre dataset</font>

#### <font color="red">PHENOMENE 1:</font>

#### L'acidité fixe, l'acidité volatile, et l'acidité citrique sont légèremment déséquilibrés dans leurs distributions et influencées par leurs maximums qui empêchent l'équilibre total. Par exemple l'acidité fixe dont la moyenne est <font color="red">6,85</font>  montre une léger déséquilibre de <font color="red">0,05</font> par rapport à la médiane.

```python
#On remarque c'est une variable continue
data["fixed acidity"].unique()
```

```python
#On utilise l'histogramme et la courbe de densité qui nous permettent de visualiser la distribution des données sur des variables continues
plt.figure(figsize=(8,5))
sns.histplot(data["fixed acidity"],kde=True)
plt.title("Distribution de l'Acide Fixe")
plt.show()
```

```python
#On vérifie sa légère asymétrie avec la skewness, en remarquant que notre valeur est légèrement supérieure à 0 et donc que son déséquillibre est situé du côté droit. 
data["fixed acidity"].skew()
```

#### <font color="red">PHENOMENE 2:</font>

#### Les substances de dioxyde de soufre libre et le total de dioxyde de soufre sont très dispersées de leurs moyennes de substances admises pour l'élaboration du vin, et ceci est certainement dû à leurs valeurs minimales qui devraient recevoir encore plus de substances et se rapprocher de la moyenne. C'est le cas du dioxyde de soufre libre dont l'écart-type <font color="red">17,007</font> est <font color="red">1/2</font> fois éloignée de sa moyenne.

```python
#Un point du diagramme de dispersion montrant l'écart-type se détâcher de la moyenne, dépassant même les 16,25 écart plus proche de la moyenne sur la figure. 
plt.figure(figsize=(8,3))
moyenne=data['free sulfur dioxide'].mean()
ecart_type=data['free sulfur dioxide'].std()
sns.scatterplot(x=[moyenne],y=[ecart_type],color='blue',label='Dispersion')
plt.xlabel('Moyenne du dioxyde de soufre libre')
plt.ylabel('Ecart-type du dioxyde de soufre libre')
plt.title('Dispersion entre Moyenne et Ecart-type')
plt.legend()
plt.show()
```

#### <font color="red">PHENOMENE 3:</font>

#### Le pH, le Sulfate, l'alcool nous transmettent comme information qu'il n'y a pas une grande variabilité des entre les substances situées à la médiane et les deux autres Quartiles donc le <font color="red">25%</font> et <font color="red">75%</font>. Nous pouvons remarquer cela avec le pH qui a une faible variabilité de <font color="red">(3,28-3,09=0,19)</font> au niveau de la médiane.

```python
# Boîte à moustache montrant en couleur bleu la petite variété de valeur entre le Q1 et Q2, sur un max et un min du PH pas très grand ce qui est logique.
plt.figure(figsize=(8,3))
sns.boxplot(x=data['pH'])
plt.title('Variabilité des données entre le Q1 et Q3')
plt.show()
```
